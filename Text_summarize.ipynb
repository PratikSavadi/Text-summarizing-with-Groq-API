StuffDocumentChain Text Summariztion

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('SQL_Notes_by_Apna_College.pdf')
docs=loader.load_and_split()
docs
```

```python
template=""" Write a concise and short summary of following speech,
speech:{text}
"""
prompt=PromptTemplate(inpt_variable=['text'],template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(model,chain_type='stuff',prompt=prompt,verbose=True)
# here when we specfic chain_type= stuff firstly it will combine the docs then passed to prompt template and then it is passed to llm model and finally
# we will get summary of text 
```

```python
output_summary=chain.run(docs)
output_summary
```

Map Reduce Technique

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
final_docs=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100).split_documents(docs)
final_docs
```

```python
# this prompt we have craeted is for the chunks means 1st chunk will give 1st summary,2nd chunk will give 2nd summary and so on...
chunk_prompt="""
Please summarize the speech
speech:'{text}'
Summary:
"""
map_prompt_template=PromptTemplate(
    input_variables=('text'),
    template=chunk_prompt
)
map_prompt_template
```

```python
final_prompt=""" Provide the final summary of entire speech
speech:'{text}'
"""
final_prompt_template=PromptTemplate(
    input_variables=('text'),
    template=final_prompt
)
final_prompt_template
```

```python
# map_prompt_template this will give summary of different chunks 
# then we will combine all the different summaries and will give final summary i.e final_prompt_template
summary_chain=load_summarize_chain(model,chain_type='map_reduce',map_prompt=map_prompt_template,combine_prompt=final_prompt_template,verbose=True)
output_res=summary_chain.run(final_docs)
output_res
```
